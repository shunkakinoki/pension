```python
low = load_workbook('data/fiscal_review.xlsx')["低位"]
middle = load_workbook('data/fiscal_review.xlsx')["中位"]
high = load_workbook('data/fiscal_review.xlsx')["高位"]
```

```python
def proc_ws(ws):
    df1 = pd.DataFrame(ws.values).loc[3:99, 0:5]
    df2 = pd.DataFrame(ws.values).loc[3:99, 7:12]
    df3 = pd.DataFrame(ws.values).loc[3:99, 14:19]
    df4 = pd.DataFrame(ws.values).loc[3:99, 21:26]
    df5 = pd.DataFrame(ws.values).loc[3:99, 28:33]
    df6 = pd.DataFrame(ws.values).loc[3:99, 35:40]

    columns = ["年度", "収入合計", "支出合計", "年度末積立金", "積立度合", "所得代替率"]
    
    df1.columns = columns
    df2.columns = columns
    df3.columns = columns
    df4.columns = columns
    df5.columns = columns
    df6.columns = columns

    return df1, df2, df3, df4, df5, df6
```

```python
low_1, low_2, low_3, low_4, low_5, low_6 = proc_ws(low)
middle_1, middle_2, middle_3, middle_4, middle_5, middle_6 = proc_ws(middle)
high_1, high_2, high_3, high_4, high_5, high_6 = proc_ws(high)
```

```python
def draw_figure_graph(name, out, x_label, y_label, df1, df2, df3, df4, df5, df6): 
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=df1[x_label], y=df1[y_label], name='ケース1',
                             line=dict(color='mediumpurple', width=3),
                             hovertext=(df1[x_label],df1[y_label])))
    fig.add_trace(go.Scatter(x=df2[x_label], y=df2[y_label], name='ケース2',
                             line=dict(color='mediumseagreen', width=3),
                             hovertext=(df2[x_label],df2[y_label])))
    fig.add_trace(go.Scatter(x=df3[x_label], y=df3[y_label], name='ケース3',
                             line=dict(color='mediumslateblue', width=3),
                             hovertext=(df3[x_label],df3[y_label])))
    fig.add_trace(go.Scatter(x=df4[x_label], y=df4[y_label], name='ケース4',
                             line=dict(color='mediumspringgreen', width=3),
                             hovertext=(df4[x_label],df4[y_label])))
    fig.add_trace(go.Scatter(x=df5[x_label], y=df5[y_label], name='ケース5',
                             line=dict(color='mediumturquoise', width=3),
                             hovertext=(df5[x_label],df5[y_label])))
    fig.add_trace(go.Scatter(x=df6[x_label], y=df6[y_label], name='ケース6',
                             line=dict(color='mediumvioletred', width=3),
                             hovertext=(df6[x_label],df6[y_label])))

    fig.update_layout(title=f'-{name}- {y_label}の推移の推移の折れ線グラフ',
                       xaxis_title=x_label,
                       yaxis_title=y_label)
    fig.update_layout(xaxis_range=['2019', '2115'],
                     xaxis_rangeslider_visible=True)
    fig.update_xaxes(tickangle=45)
    fig.write_html(f'output/review/{out}.html', auto_open=False)
    fig.write_image(f'output/review/{out}.png')
    fig.write_image(f'output/review/{out}.jpg')
#     fig.show()
```

```python
draw_figure_graph('出生率低位', 'low_reserve', '年度', '積立度合', low_1, low_2, low_3, low_4, low_5, low_6)
draw_figure_graph('出生率低位', 'low_expense', '年度', '支出合計', low_1, low_2, low_3, low_4, low_5, low_6)
draw_figure_graph('出生率低位', 'low_revenue', '年度', '収入合計', low_1, low_2, low_3, low_4, low_5, low_6)

draw_figure_graph('出生率中位', 'middle_reserve', '年度', '積立度合', low_1, low_2, low_3, low_4, low_5, low_6)
draw_figure_graph('出生率中位', 'middle_expense', '年度', '支出合計', low_1, low_2, low_3, low_4, low_5, low_6)
draw_figure_graph('出生率中位', 'middle_revenue', '年度', '収入合計', low_1, low_2, low_3, low_4, low_5, low_6)

draw_figure_graph('出生率高位', 'high_reserve', '年度', '積立度合', low_1, low_2, low_3, low_4, low_5, low_6)
draw_figure_graph('出生率高位', 'high_expense', '年度', '支出合計', low_1, low_2, low_3, low_4, low_5, low_6)
draw_figure_graph('出生率高位', 'high_revenue', '年度', '収入合計', low_1, low_2, low_3, low_4, low_5, low_6)
```
