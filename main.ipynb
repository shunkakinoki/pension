```python
import pandas as pd
from IPython.display import display
from openpyxl import load_workbook

import plotly.express as px
from plotly.subplots import make_subplots
import plotly.figure_factory as ff
import plotly.graph_objects as go
import plotly.io as pio
```

```python
a = load_workbook('data/a.xlsx')["国民年金"]
b = load_workbook('data/b.xlsx')["国民年金"]
c = load_workbook('data/c.xlsx')["国民年金"]
d = load_workbook('data/d.xlsx')["国民年金"]
e = load_workbook('data/e.xlsx')["国民年金"]
f = load_workbook('data/f.xlsx')["国民年金"]
```

```python
def proc_ws(ws):
    df = pd.DataFrame(ws.values).loc[20:112, [1,3,8,13]]
    df.columns = ["年度", "収入合計", "支出合計", "年度末積立金"]
    return df
```

```python
aws = proc_ws(a)
bws = proc_ws(b)
cws = proc_ws(c)
dws = proc_ws(d)
ews = proc_ws(e)
fws = proc_ws(f)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=aws['年度'], y=aws['年度末積立金'], name='ケースA',
                         line=dict(color='mediumpurple', width=3),
                         hovertext=(aws['年度'],aws['年度末積立金'])))
fig.add_trace(go.Scatter(x=bws['年度'], y=bws['年度末積立金'], name='ケースB',
                         line=dict(color='mediumseagreen', width=3),
                         hovertext=(bws['年度'],bws['年度末積立金'])))
fig.add_trace(go.Scatter(x=cws['年度'], y=cws['年度末積立金'], name='ケースC',
                         line=dict(color='mediumslateblue', width=3),
                         hovertext=(cws['年度'],cws['年度末積立金'])))
fig.add_trace(go.Scatter(x=dws['年度'], y=dws['年度末積立金'], name='ケースD',
                         line=dict(color='mediumspringgreen', width=3),
                         hovertext=(cws['年度'],dws['年度末積立金'])))
fig.add_trace(go.Scatter(x=ews['年度'], y=ews['年度末積立金'], name='ケースE',
                         line=dict(color='mediumturquoise', width=3),
                         hovertext=(cws['年度'],ews['年度末積立金'])))
fig.add_trace(go.Scatter(x=fws['年度'], y=fws['年度末積立金'], name='ケースF',
                         line=dict(color='mediumvioletred', width=3),
                         hovertext=(cws['年度'],fws['年度末積立金'])))

fig.update_layout(title='年度末積立金の推移の推移の折れ線グラフ',
                   xaxis_title='年度',
                   yaxis_title='年度末積立金')
fig.update_layout(xaxis_range=['2016', '2110'],
                 xaxis_rangeslider_visible=True)
fig.update_xaxes(tickangle=45)
fig.write_html(f'output/deposit_all.html', auto_open=False)
fig.show()
```
